```python
import matplotlib.pyplot as plt
import numpy as np
from paper_hll import simulation1
```

```python
A = 10**4
r = 0.1
B = round(A*r)
m = 100
p = 1/m
k = 10
ntrial = 100

exp1, estimates1 = simulation1(A,m,r,ntrial,k)
print(exp1)
print(estimates1)


# Create the histogram
plt.hist(estimates1, bins=30)

# Add titles and labels
plt.title('Histogram of Random Data, without correcting factor')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Find the minimum and maximum values across both datasets
min_val = min(min(estimates1), min(estimates2))
max_val = max(max(estimates1), max(estimates2))

# Define bin edges based on the same min and max values
bin_edges = np.linspace(min_val, max_val, 30 + 1)  # +1 because linspace gives bin edges, not bin centers

plt.hist(estimates6, bins=bin_edges, alpha=0.2, label="995 matches", density=True)
plt.hist(estimates5, bins=bin_edges, alpha=0.2, label="996 matches", density=True)
plt.hist(estimates4, bins=bin_edges, alpha=0.2, label="997 matches", density=True)
plt.hist(estimates3, bins=bin_edges, alpha=0.2, label="998 matches", density=True)
plt.hist(estimates2, bins=bin_edges, alpha=0.2, label="999 matches", density=True)
plt.hist(estimates1, bins=bin_edges, alpha=0.2, label="1000 matches", density=True)


# Add titles and labels
plt.title('Histogram of Random Data, without correcting factor')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Display the plot
plt.show()
```
